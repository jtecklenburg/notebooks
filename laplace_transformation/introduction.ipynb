# Numerical inverse Laplace transformation

Numerical inverse [Laplace transformation](https://en.wikipedia.org/wiki/Laplace_transform) is possible!

Let us try an rectangular impulse defined in Laplace domain as

$$F(s) = \frac{1}{s}\left(1-e^{-\tau\cdot s}\right)$$

Here we define $\tau = 1$

```python
from invlaplace import crump, iseger
from numpy import exp, arange
import matplotlib.pyplot as plt

def L(s): return 1/s*1/(1+exp(-s))

dt = 0.01
T = 20 

t = arange(0, T, dt)

ft1 = crump(L, t)
ft2 = iseger(L, t)

plt.plot(t, ft1, t, ft2)
plt.xlabel('t')
plt.ylabel('f(t)')
plt.legend(['Crump', 'Den Iseger'])
plt.show()
```

Strong gradients seems to introduce oscillations, but it looks like a rectangular impulse. 

This time we convolute a delayed unit step with an exponential decay. In Laplace domain a convolution is a simple multiplication. 

$$L(s) = \frac{1}{s}e^{-5\cdot s} \cdot \frac{1}{s + 0.5} $$

```python
def L(s): return 1/s*(exp(-5*s))*1/(0.5+s)

ft1 = crump(L, t)
ft2 = iseger(L, t)

plt.plot(t, ft1, t, ft2)
plt.xlabel('t')
plt.ylabel('f(t)')
plt.legend(['Crump', 'Den Iseger'])
plt.show()
```

No strong gradients, no oscillations.
