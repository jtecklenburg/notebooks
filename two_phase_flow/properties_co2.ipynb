# Thermophysical properties of carbon dioxide

The [NIST Chemistry WebBook](https://webbook.nist.gov/chemistry/fluid/) provides thermophysical data for several substances. Here is compilation of thermopysical data for carbon dioxide from NIST in the Excel file "Properties_CO.xlsx".

```python
import properties

rho = properties.from_table("Density (kg/m3)")
```

```python
mu = properties.from_table("Viscosity (uPa*s)")
```

```python
phase = properties.from_table("Phase2")
```

Here -1 indicates the gaseous state, 1 indicates the liquid state and 0 the supercritical state.

When we want to know the properties for 60C and 80 MPa. The function `properties.from_table()' interpolates the values from the Excel file. This approach may be inaccurate in region with large gradients.

```python
T = 60
p = 80

print(f"The density of carbon dioxide is {rho(p, T)} kg/m for a temperature of {T}C and a pressure of {p}MPa.")

print(f"The viscosity of carbon dioxide is {mu(p, T)} uPa*s for a temperature of {T}C and a pressure of {p}MPa.")

phasedict = {-1: "gaseous", 0: "supercritical", 1: "liquid"}
print(f"Carbon dioxide is {phasedict[int(phase(60, 80))]} for a temperature of {T}C and a pressure of {p}MPa.")
```
